```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python

act = pd.read_csv("OTg6QzA_activities.csv")
print(act.shape)
print(act.head())
```

```python
X = act.X.to_numpy()
Y = act.Y.to_numpy()

print(X)
print(Y)
```

```python
plt.style.use('grayscale')
plt.figure(facecolor='white')

plt.scatter(X, Y)
plt.show()
```

```python
plt.figure(facecolor='white')
plt.hist2d(X, Y, bins=(25,25), cmap = plt.cm.Greys)
cb = plt.colorbar()
cb.set_label('frequency of appearance, bin=25')
plt.show()

plt.figure(facecolor='white')
plt.hist2d(X, Y, bins=(40,40), cmap = plt.cm.Greys)
cb = plt.colorbar()
cb.set_label('frequency of appearance, bin=40')
plt.show()

plt.figure(facecolor='white')
plt.hexbin(X, Y, gridsize=20, cmap=plt.cm.Greys)
cb = plt.colorbar()
cb.set_label('frequency of appearance, gridsize=20')
plt.show()
```

```python
x1 = X[:16570]
x2 = X[16570:16570*2]
x3 = X[16570*2:16570*3]
x4 = X[16570*3:]

y1 = Y[:16570]
y2 = Y[16570:16570*2]
y3 = Y[16570*2:16570*3]
y4 = Y[16570*3:]

plt.figure(facecolor='white')

# 1
plt.subplot(2,2,1)
#plt.hist2d(x1, y1, bins=20, cmap='viridis')
plt.scatter(x1, y1)
plt.title('1-Q')
plt.colorbar()

# 2
plt.subplot(2,2,2)
#plt.hist2d(x2, y2, bins=20, cmap='plasma')
plt.scatter(x2, y2)
plt.title('2-Q')
plt.colorbar()

# 3
plt.subplot(2,2,3)
#plt.hist2d(x3, y3, bins=20, cmap='inferno')
plt.scatter(x3, y3)
plt.title('3-Q')
plt.colorbar()

# 4
plt.subplot(2,2,4)
#plt.hist2d(x4, y4, bins=20, cmap='magma')
plt.scatter(x4, y4)
plt.title('4-Q')
plt.colorbar()
```

```python
plt.figure(facecolor='white')

plt.scatter(x1, y1, c='r')
plt.scatter(x2, y2, c='b')
plt.scatter(x3, y3, c='k')
plt.scatter(x4, y4, c='y')
plt.title('ALL')
plt.grid()

plt.show()
```
